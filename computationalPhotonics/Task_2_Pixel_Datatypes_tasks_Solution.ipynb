```python
from imageio import imread
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
```

# 2. Bit (Pixel) Depth

## Task 2.1
Execute the code beneath and try to answer following questions:

- What influence does the bit depth have on the quality of a digital image? <br>
     <font color=blue>
      The represents the number of bits to represent the intensity of an
      image.<br> A high bit depth has a higher the number of values or shades
      <br>of colors to represent a digital image and the better.
     </font>
     
     
- How can we change the bit depth of an image in python? <br>
    <font color=blue> 
     You call the 'astype()' method in numpy on the image and chose a <br>
     pixel type to convert to as used in Task 2.2
    </font>
    
    
- What is crucial if we change from integers to floating point pixel values and vice versa?<br>
     <font color=blue> 
     When changed from integers to floating points the size is <br>
     increased because higher bitdepth is used to represent the <br>
     image. On the other way round information is lost when image <br>
     change from floating points to integer.
    </font>
    

```python
im = imread("../imageProcessing/samples/lena.jpg", pilmode="L")
# Down-scale an image
# the resulting pixel type is double precision float (float64)
im2 = resize(im, [256, 256])

# convert pixel type to unsigned 8-bit integer
im3 = im2.astype(np.uint8)

# Display information about the image size, pixel type and how many bytes the image has
fmt_string = "Size: {} \t Data Type: {} \t #bytes: {:,} \t kB: {:,}"
print(fmt_string.format(im.shape,   im.dtype,  im.nbytes, im.nbytes / 1024))
print(fmt_string.format(im2.shape,  im2.dtype, im2.nbytes, im2.nbytes / 1024))
print(fmt_string.format(im3.shape,  im3.dtype, im3.nbytes, im3.nbytes / 1024))
```

```python
# visualize all 3 images
fig, axs = plt.subplots(1,3, figsize=(16,9))

axs[0].imshow(im, cmap=plt.cm.gray)
axs[0].set_title("Original image")

axs[1].imshow(im2, cmap=plt.cm.gray)
axs[1].set_title("Downscaled image")

axs[2].imshow(im3, cmap=plt.cm.gray)
axs[2].set_title("Downscaled image with unsigned 8-bit integer pixel values")

plt.show()
```

## Task 2.2
How can we *"correct"* the negative image `samples/negative.jpg`?

<img src="../../samples/negative.jpg">

```python
im = imread("../imageProcessing/samples/negative.jpg", pilmode="L")

im_corrected = -im

fig, axs = plt.subplots(1,2, figsize=(16,9))

axs[0].imshow(im, cmap=plt.cm.gray)
axs[0].set_title("Original image")

axs[1].imshow(im_corrected, cmap=plt.cm.gray)
axs[1].set_title("Corrected image")
plt.show()
```
