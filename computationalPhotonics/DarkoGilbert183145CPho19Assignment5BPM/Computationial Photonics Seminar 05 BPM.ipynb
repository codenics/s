```python
import numpy as np
import scipy.sparse as sps
from scipy.sparse.linalg import eigs


def guided_modes_1DTE(prm, k0, h):
    """Computes the effective permittivity of a TE polarized guided eigenmode.
    All dimensions are in µm.
    Note that modes are filtered to match the requirement that
    their effective permittivity is larger than the substrate (cladding).

    Parameters
    ----------
    prm : 1d-array
        Dielectric permittivity in the x-direction
    k0 : float
        Free space wavenumber
    h : float
        Spatial discretization

    Returns
    -------
    eff_eps : 1d-array
        Effective permittivity vector of calculated modes
    guided : 2d-array
        Field distributions of the guided eigenmodes
    """

    # set up operator matrix
    main = -2.0*np.ones((len(prm),))/(h**2*k0**2) + prm
    sec = np.ones((len(prm)-1,))/(h**2*k0**2)
    L = np.diag(sec,-1) + np.diag(main,0) + np.diag(sec,1);

    # solve eigenvalue problem
    eff_eps, guided = np.linalg.eig(L);

    # pick only guided modes
    ind = eff_eps > max(prm[0], prm[-1]);
    eff_eps = eff_eps[ind];
    guided  = guided[:,ind];

    # sort modes from highest to lowest effective permittivity
    # (the fundamental mode has the highest effective permittivity)
    I = np.argsort(-eff_eps);
    eff_eps = eff_eps[I]
    guided = guided[:,I]
    return eff_eps, guided
```

```python
def guided_modes_2D(prm, k0, h, numb):
    """Computes the effective permittivity of a quasi-TE polarized guided
    eigenmode. All dimensions are in µm.

    Parameters
    ----------
    prm  : 2d-array
        Dielectric permittivity in the xy-plane
    k0 : float
        Free space wavenumber
    h : float
        Spatial discretization
    numb : int
        Number of eigenmodes to be calculated

    Returns
    -------
    eff_eps : 1d-array
        Effective permittivity vector of calculated eigenmodes
    guided : 3d-array
        Field distributions of the guided eigenmodes
    """

    NX, NY = prm.shape
    prm = prm*k0**2
    N = NX*NY
    ihx2 = 1/h**2
    ihy2 = 1/h**2

    md = -2.0*(ihx2 + ihy2)*np.ones((N,)) + prm.ravel(order='F')
    xd = ihx2*np.ones((N,))
    yd = ihy2*np.ones((N,))
    H  = sps.spdiags([yd,xd,md,xd,yd], [-NX,-1,0,1,NX], N, N, format='csc')

    #remove the '1' when moving to a new line
    # -> look into script pg. 31, lower blue line in first figure
    for i in range(1, NY):
        n = i*NX
        H[n-1, n] = 0
        H[n, n-1] = 0

    # solve eigenvalue problem
    eigvals, eigvecs = eigs(H, k=numb, which='LR', maxiter=3000)
    eff_eps = eigvals/(k0**2);

    # pick only guided modes
    eps_clad = max([prm[:,0].max(), prm[:,-1].max(),
                    prm[0,:].max(), prm[-1,:].max()])
    ind = eff_eps > eps_clad/k0**2
    eff_eps = eff_eps[ind]
    eigvecs = eigvecs[:,ind]

    # sort modes from highest to lowest effective permittivity
    # (the fundamental mode has the highest effective permittivity)
    I = np.argsort(-eff_eps);
    eff_eps = eff_eps[I]
    eigvecs = eigvecs[:,I]

    # reshape eigenvectors to a 2D matrix and store them in a 3D array
    guided = np.zeros((len(eff_eps),NX,NY), dtype=eigvecs.dtype)
    for i in range(len(eff_eps)):
       guided[i,:,:] = np.reshape(eigvecs[:,i], (NX,NY), order='F')

    return eff_eps, guided
```

```python
import numpy as np
import scipy.sparse.linalg
import scipy.sparse as sps


def waveguide(xa, xb, Nx, n_cladding, n_core):
    '''Generates the refractive index distribution of a slab waveguide
    with step profile centered around the origin of the coordinate
    system with a refractive index of n_core in the waveguide region
    and n_cladding in the surrounding cladding area.
    All lengths have to be specified in µm.

    Parameters
    ----------
        xa : float
            Width of calculation window
        xb : float
            Width of waveguide
        Nx : int
            Number of grid points
        n_cladding : float
            Refractive index of cladding
        n_core : float
            Refractive index of core

    Returns
    -------
        n : 1d-array
            Generated refractive index distribution
        x : 1d-array
            Generated coordinate vector
    '''
    x = np.linspace(-0.5*xa, 0.5*xa, Nx)
    n = np.ones_like(x)*n_cladding
    n[np.abs(x) <= 0.5*xb] = n_core
    return n, x


def gauss(xa, Nx, w):
    '''Generates a Gaussian field distribution v = exp(-x^2/w^2) centered
    around the origin of the coordinate system and having a width of w.
    All lengths have to be specified in µm.

    Parameters
    ----------
        xa : float
            Width of calculation window
        Nx : int
            Number of grid points
        w  : float
            Width of Gaussian field

    Returns
    -------
        v : 1d-array
            Generated field distribution
        x : 1d-array
            Generated coordinate vector
    '''
    x = np.linspace(-0.5*xa, 0.5*xa, Nx)
    v = np.exp(-(x/w)**2)
    return v, x


def bpm_operator(dx, n, nd, lam):
    '''Calculates the finite difference discretization of the
    BPM differential operator L = 1i/(2*kbar)*(d^2/dx^2 + k^2 - kbar^2).
    All lengths have to be specified in µm.

    Parameters
    ----------
        dx : float
            Lateral step size
        n : 1d-array
            Refractive index distribution
        nd : float
            Reference refractive index
        lam: float
            Wavelength

    Returns
    -------
        L : sparse matrix
            Beam propagation operator
    '''
    Nx    = n.size
    k     = 2*np.pi/lam*n
    kbar  = 2*np.pi/lam*nd
    alpha = 1/((dx**2)*2*kbar)  # abbreviation for prefactor
    Wi    = (k**2 - kbar**2)/(2*kbar) # abbreviation for last term

    main_diag = 1j*(Wi - 2*alpha)
    aux_diag  = 1j*alpha*np.ones((Nx,))
    L = sps.spdiags([aux_diag, main_diag, aux_diag], [-1, 0, 1], Nx, Nx)
    return L


def beamprop_explicit(v_in, lam, dx, n, nd, z_end, dz, output_step):
    '''Propagates an initial field over a given distance based on the
    solution of the paraxial wave equation in an inhomogeneous
    refractive index distribution using the explicit
    forward Euler scheme. All lengths have to be specified in µm.

    Parameters
    ----------
        v_in : 1d-array
            Initial field
        lam : float
            Wavelength
        dx : float
            Transverse step size
        n : 1d-array
            Refractive index distribution
        nd : float
            Reference refractive index
        z_end : float
            Propagation distance
        dz : float
            Step size in propagation direction
        output_step : int
            Number of steps between field outputs

    Returns
    -------
        v_out : 2d-array
            Propagated field
        z : 1d-array
            z-coordinates of field output
    '''
    # calculate grid size
    Nx = n.size
    Nz = int(np.round(z_end/(output_step*dz))) + 1 # +1 for initial field at z = 0
    # calculate required number of iterations
    Niter = int((Nz - 1)*output_step)

    # generate the output grid
    v_out = np.zeros((Nx, Nz), dtype=np.complex128)
    z = np.arange(Nz)*dz*output_step

    # generate iteration matrix A = I + dz*L
    A = sps.eye(Nx) + dz*bpm_operator(dx, n, nd, lam)

    # iteration
    curr_slice = 0 # current postion in output matrix
    v_curr = v_in[:,np.newaxis] # current field
    v_out[:,curr_slice] = v_curr.flat # store input field as first output
    for i in range(Niter):
        # advance field: multiply current field with iteration matrix
        v_curr = A.dot(v_curr)
        # check if current iteration has to be stored
        if (i % output_step) == 0:
            curr_slice += 1 # move to next output slot
            v_out[:,curr_slice] = v_curr.flat # store current solution
    return v_out, z

'''Test script for task I (explicit scheme) - BPM.
'''

import numpy as np
from matplotlib import pyplot as plt
import time

# mode solver for comparison with waveguide mode

# dark bluered colormap, registers automatically with matplotlib on import
import bluered_dark

plt.rcParams.update({
        'figure.figsize': (12/2.54, 9/2.54),
        'figure.subplot.bottom': 0.15,
        'figure.subplot.left': 0.165,
        'figure.subplot.right': 0.925,
        'figure.subplot.top': 0.9,
        'axes.grid': True,
})
plt.close('all')
save_figures = False

# %%
# computational parameters
z_end   = 100       # propagation distance
lam     = 1         # wavelength
nd      = 1.455     # referecne index
xa      = 50        # size of computational window
Nx      = 251       # number of transverse points
dx      = xa/(Nx-1) # transverse step size

# waveguide parameters
xb      = 2.0       # size of waveguide
n_clad  = 1.45      # cladding index
n_core  = 1.46      # core refr. index

# source width
w       = 5.0       # Gaussian beam width

# propagation step size
dz = 0.0025
output_step = int(np.round(1.0/dz))

# create index distribution
n, x = waveguide(xa, xb, Nx, n_clad, n_core)

# create initial field
v_in, x = gauss(xa, Nx, w)
v_in = v_in/np.sqrt(np.sum(np.abs(v_in)**2)) # normalize power to unity

# calculate fundamental waveguide mode for comparison
eff_eps, guided = guided_modes_1DTE(n**2,2*np.pi/lam, dx)
v_ref = guided[:,0]
ind = np.argmax(np.abs(v_ref)) # find index of max
v_ref = v_ref*np.exp(-1j*np.angle(v_ref[ind])) # set phase of max to zero
v_ref = v_ref/np.sqrt(np.sum(np.abs(v_ref)**2)) # normalize power to unity

overlap = np.sum(v_in*np.conj(v_ref))
print('Fundamental waveguide mode:\n'
      '\tn_eff = {0:g}\n\toverlapp = {1:g}'.format(
              np.sqrt(eff_eps[0]), np.abs(overlap)))

# %% plot initial field, waveguide mode and index distribution
delta_n = n_core - n_clad
fig = plt.figure()
plt.plot(x, np.real(v_in)/np.max(np.abs(v_ref))*delta_n,
         label='input field')
plt.plot(x, np.real(v_ref)/np.max(np.abs(v_ref))*delta_n,
         label='fundamental\nwaveguide mode')
plt.plot(x, n - n_clad,'k-',
         label='index difference')
plt.xlim(x[[0, -1]])
plt.ylim(np.array([-0.05, 1.05])*(n_core - n_clad))
plt.grid(True)
plt.xlabel('lateral position [µm]')
plt.ylabel(' ')
plt.legend(frameon=False)
plt.show()

if save_figures:
    fig.savefig('waveguide.pdf', dpi=600)

# %% propagate field
t = time.time()
v_out_ex, z = beamprop_explicit(v_in, lam, dx, n, nd,
                                  z_end, dz, output_step)
print('elapsed time: {0:g}'.format(time.time() - t))

fig = plt.figure()
v = np.abs(v_out_ex).max()
plt.contourf(z, x, v_out_ex.real, np.linspace(-v , v, 101),
             cmap='bluered_dark', zorder=-5)
plt.gca().set_rasterization_zorder(-1)
cb = plt.colorbar()
plt.xlabel('propagation distance [µm]')
plt.ylabel('lateral position [µm]')
plt.title('explicit FTCS scheme')
cb.set_label('real part of field')
cb.set_ticks(np.linspace(-0.3, 0.3, 7))
plt.show()

if save_figures:
    fig.savefig('BPM_explicit.pdf', dpi=300)

# %% influence of step size and lateral resolution
fig = plt.figure()

for i, (dz, Nx) in enumerate([(0.0025, 251), (0.005, 251), (0.0025, 301)]):
    dx = xa/(Nx - 1)
    output_step = int(np.round(1.0/dz))
    n, x    = waveguide(xa, xb, Nx, n_clad, n_core)
    v_in, x = gauss(xa, Nx, w)
    v_in= v_in/np.sqrt(np.sum(np.abs(v_in)**2))
    v_out_ex, z = beamprop_explicit(v_in, lam, dx, n, nd,
                                    z_end, dz, output_step)
    plt.plot(x, v_out_ex[:,-1].real + 0.1*i,
             label='dz = {0:1.1f}nm, Nx={1}'.format(dz*1e3, Nx))

plt.grid(True)
plt.legend(frameon=False)
plt.xlim(x[[0, -1]])
plt.xlabel('lateral position [µm]')
plt.ylabel('real part of field afer 100 µm')
plt.show()

if save_figures:
    fig.savefig('BPM_explicit_parameters.pdf', dpi=600)



```

```python
def beamprop_implicit(v_in, lam, dx, n, nd, z_end, dz, output_step):
    '''Propagates an initial field over a given distance based on the
    solution of the paraxial wave equation in an inhomogeneous
    refractive index distribution using the explicit
    backward Euler scheme. All lengths have to be specified in µm.

    Parameters
    ----------
        v_in : 1d-array
            Initial field
        lam : float
            Wavelength
        dx : float
            Transverse step size
        n : 1d-array
            Refractive index distribution
        nd : float
            Reference refractive index
        z_end : float
            Propagation distance
        dz : float
            Step size in propagation direction
        output_step : int
            Number of steps between field outputs

    Returns
    -------
        v_out : 2d-array
            Propagated field
        z : 1d-array
            z-coordinates of field output
    '''
    # calculate grid size
    Nx = n.size
    Nz = int(np.round(z_end/(output_step*dz))) + 1 # +1 for initial field at z = 0
    # calculate required number of iterations
    Niter = int((Nz - 1)*output_step)

    # generate the output grid
    v_out = np.zeros((Nx, Nz), dtype=np.complex128)
    z = np.arange(Nz)*dz*output_step

    # generate iteration matrix A = I - dz*L
    A = sps.eye(Nx) - dz*bpm_operator(dx, n, nd, lam)

    # iteration
    curr_slice = 0 # current postion in output matrix
    v_curr = v_in[:,np.newaxis] # current field
    v_out[:,curr_slice] = v_curr.flat # store input field as first output
    for i in range(Niter):
        # advance field: solve tridiagonal system of equations
        v_curr = sps.linalg.spsolve(A, v_curr)
        # check if current iteration has to be stored
        if (i % output_step) == 0:
            curr_slice += 1 # move to next output slot
            v_out[:,curr_slice] = v_curr.flat # store current solution
    return v_out, z

'''Test script for task II (implicit scheme) of the assignment of - BPM.
'''

import numpy as np
from matplotlib import pyplot as plt
import time

# mode solver for comparison with waveguide mode

# dark bluered colormap, registers automatically with matplotlib on import
import bluered_dark

plt.rcParams.update({
        'figure.figsize': (12/2.54, 9/2.54),
        'figure.subplot.bottom': 0.15,
        'figure.subplot.left': 0.165,
        'figure.subplot.right': 0.925,
        'figure.subplot.top': 0.9,
        'axes.grid': True,
})
plt.close('all')
save_figures = False

# %%
# computational parameters
z_end   = 100       # propagation distance
lam     = 1         # wavelength
nd      = 1.455     # referecne index
xa      = 50        # size of computational window
Nx      = 251       # number of transverse points
dx      = xa/(Nx-1) # transverse step size

# waveguide parameters
xb      = 2.0       # size of waveguide
n_clad  = 1.45      # cladding index
n_core  = 1.46      # core refr. index

# source width
w       = 5.0       # Gaussian beam width

# propagation step size
dz = 0.1
output_step = np.round(1.0/dz)

# create index distribution
n, x = waveguide(xa, xb, Nx, n_clad, n_core)

# create initial field
v_in, x = gauss(xa, Nx, w)
v_in = v_in/np.sqrt(np.sum(np.abs(v_in)**2)) # normalize power to unity

# calculate fundamental waveguide mode for comparison
eff_eps, guided = guided_modes_1DTE(n**2,2*np.pi/lam, dx)
v_ref = guided[:,0]
ind = np.argmax(np.abs(v_ref)) # find index of max
v_ref = v_ref*np.exp(-1j*np.angle(v_ref[ind])) # set phase of max to zero
v_ref = v_ref/np.sqrt(np.sum(np.abs(v_ref)**2)) # normalize power to unity

overlap = np.sum(v_in*np.conj(v_ref))
print('Fundamental waveguide mode:\n'
      '\tn_eff = {0:g}\n\toverlapp = {1:g}'.format(
              np.sqrt(eff_eps[0]), np.abs(overlap)))

# %% propagate field
t = time.time()
v_out_im, z = beamprop_implicit(v_in, lam, dx, n, nd,
                                  z_end, dz, output_step)
print('elapsed time: {0:g}'.format(time.time() - t))

fig = plt.figure()
v = np.abs(v_out_im).max()
plt.contourf(z, x, v_out_im.real, np.linspace(-v , v, 101),
             cmap='bluered_dark', zorder=-5)
plt.gca().set_rasterization_zorder(-1)
cb = plt.colorbar()
plt.xlabel('propagation distance [µm]')
plt.ylabel('lateral position [µm]')
plt.title('implicit BTCS scheme')
cb.set_label('real part of field')
cb.set_ticks(np.linspace(-0.3, 0.3, 7))
plt.show()

if save_figures:
    fig.savefig('BPM_implicit.pdf', dpi=300)

# %% comparison with waveguide mode mode
overlap = np.sum(v_out_im[:,-1]*np.conj(v_ref)) # overlap with waveguide mode
print('overlapp with waveguide mode: {0:g}'.format(np.abs(overlap)))
fig = plt.figure()
plt.plot(x, np.real(v_in), label='initial field')
plt.plot(x, np.real(v_out_im[:,-1]), label='propagated field\nafter 100µm')
plt.plot(x, np.real(overlap*v_ref), label='waveguide mode\n(scaled)')
plt.grid(True)
plt.xlim(x[[0, -1]])
plt.ylim(np.array([-0.25, 1.05])*np.max(np.abs(v_out_im[:,-1])))
plt.legend(frameon=False)
plt.xlabel('lateral position [µm]')
plt.ylabel('real part of field')
plt.show()

if save_figures:
    fig.savefig('BPM_implicit_field_at_100mum.pdf', dpi=600)

# %% propagate field (long distance)
z_end = 500
t = time.time()
v_out_im, z = beamprop_implicit(v_in, lam, dx, n, nd, z_end, dz, output_step)
print('elapsed time: {0:g}'.format(time.time() - t))

fig = plt.figure()
v = np.abs(v_out_im).max()
plt.contourf(z, x, v_out_im.real, np.linspace(-v , v, 101),
             cmap='bluered_dark', zorder=-5)
plt.gca().set_rasterization_zorder(-1)
cb = plt.colorbar()
plt.xlabel('propagation distance [µm]')
plt.ylabel('lateral position [µm]')
plt.title('implicit BTCS scheme')
cb.set_label('real part of field')
cb.set_ticks(np.linspace(-0.3, 0.3, 7))
plt.show()

if save_figures:
    fig.savefig('BPM_implicit_long_distance.pdf', dpi=300)


```

```python
def beamprop_CN(v_in, lam, dx, n, nd,  z_end, dz, output_step):
    '''Propagates an initial field over a given distance based on the
    solution of the paraxial wave equation in an inhomogeneous
    refractive index distribution using the explicit-implicit
    Crank-Nicolson scheme. All lengths have to be specified in µm.

    Parameters
    ----------
        v_in : 1d-array
            Initial field
        lam : float
            Wavelength
        dx : float
            Transverse step size
        n : 1d-array
            Refractive index distribution
        nd : float
            Reference refractive index
        z_end : float
            Propagation distance
        dz : float
            Step size in propagation direction
        output_step : int
            Number of steps between field outputs

    Returns
    -------
        v_out : 2d-array
            Propagated field
        z : 1d-array
            z-coordinates of field output
    '''
    # calculate grid size
    Nx = n.size
    Nz = int(np.round(z_end/(output_step*dz))) + 1 # +1 for initial field at z = 0
    # calculate required number of iterations
    Niter = int((Nz - 1)*output_step)

    # generate the output grid
    v_out = np.zeros((Nx, Nz), dtype=np.complex128)
    z = np.arange(Nz)*dz*output_step

    # generate iteration matrices 
    # A = I - 0.5*dz*L and B = I + 0.5*dz*L
    B = 0.5*dz*bpm_operator(dx, n, nd, lam)
    A = sps.eye(Nx) - B
    B = sps.eye(Nx) + B

    # iteration
    curr_slice = 0 # current postion in output matrix
    v_curr = v_in[:,np.newaxis] # current field
    v_out[:,curr_slice] = v_curr.flat # store input field as first output
    for i in range(Niter):
        # advance field: solve tridiagonal system of equations
        v_curr = sps.linalg.spsolve(A, B.dot(v_curr))
        # check if current iteration has to be stored
        if (i % output_step) == 0:
            curr_slice += 1 # move to next output slot
            v_out[:,curr_slice] = v_curr.flat # store current solution
    return v_out, z



'''Test script for task III (Crank-Nicolson scheme)- BPM
'''

import numpy as np
from matplotlib import pyplot as plt
import time

# mode solver for comparison with waveguide mode

# dark bluered colormap, registers automatically with matplotlib on import
import bluered_dark

plt.rcParams.update({
        'figure.figsize': (12/2.54, 9/2.54),
        'figure.subplot.bottom': 0.15,
        'figure.subplot.left': 0.165,
        'figure.subplot.right': 0.925,
        'figure.subplot.top': 0.9,
        'axes.grid': True,
})
plt.close('all')
save_figures = False

# %%
# computational parameters
z_end   = 100       # propagation distance
lam     = 1         # wavelength
nd      = 1.455     # referecne index
xa      = 50        # size of computational window
Nx      = 251       # number of transverse points
dx      = xa/(Nx-1) # transverse step size

# waveguide parameters
xb      = 2.0       # size of waveguide
n_clad  = 1.45      # cladding index
n_core  = 1.46      # core refr. index

# source width
w       = 5.0       # Gaussian beam width

# propagation step size
dz = 0.5
output_step = np.round(1.0/dz)

# create index distribution
n, x = waveguide(xa, xb, Nx, n_clad, n_core)

# create initial field
v_in, x = gauss(xa, Nx, w)
v_in = v_in/np.sqrt(np.sum(np.abs(v_in)**2)) # normalize power to unity

# calculate fundamental waveguide mode for comparison
eff_eps, guided = guided_modes_1DTE(n**2,2*np.pi/lam, dx)
v_ref = guided[:,0]
ind = np.argmax(np.abs(v_ref)) # find index of max
v_ref = v_ref*np.exp(-1j*np.angle(v_ref[ind])) # set phase of max to zero
v_ref = v_ref/np.sqrt(np.sum(np.abs(v_ref)**2)) # normalize power to unity

overlap = np.sum(v_in*np.conj(v_ref))
print('Fundamental waveguide mode:\n'
      '\tn_eff = {0:g}\n\toverlapp = {1:g}'.format(
              np.sqrt(eff_eps[0]), np.abs(overlap)))

# %% propagate field
t = time.time()
v_out_CN, z = beamprop_CN(v_in, lam, dx, n, nd,
                                  z_end, dz, output_step)
print('elapsed time: {0:g}'.format(time.time() - t))

fig = plt.figure()
v = np.abs(v_out_CN).max()
plt.contourf(z, x, v_out_CN.real, np.linspace(-v , v, 101),
             cmap='bluered_dark', zorder=-5)
plt.gca().set_rasterization_zorder(-1)
cb = plt.colorbar()
plt.xlabel('propagation distance [µm]')
plt.ylabel('lateral position [µm]')
plt.title('Crank-Nicolson scheme')
cb.set_label('real part of field')
cb.set_ticks(np.linspace(-0.3, 0.3, 7))
plt.show()

if save_figures:
    fig.savefig('BPM_CN.pdf', dpi=300)
```
