```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vv = np.arange(-10, 10, 0.1)
```

```python
vv.max(), vv.min()
```

```python
vv
```

```python
zz = vv*vv
yy = -0.7*zz + 100
```

```python
plt.plot(vv, zz)
plt.plot(vv, yy)
```

```python
vv = np.arange(0, 1, 0.001)
zz = vv*vv
yy = -0.7*zz + 1
zz = zz/zz.max()
yy = yy/yy.max()
```

```python
plt.plot(vv, zz)
plt.plot(vv, yy)
```
